```python
import numpy as np
```

```python
pts = {'a': (1,2), 'b': (2, 3), 'c': (2, 4), 'd': (6, 5), 'e': (3, 1), 'f': (3, 3), 'g': (4,4), 'h': (7,6), 'i': (8,8),
       'j': (8,9), 'k': (9,4), 'l':(5,9), 'm':(4,7), 'n':(6,8), 'o':(6,4)}

cluster_centers = [(2,3), (6,2), (8,8)]
```

```python
for pt_name, pt in pts.items():
    for cluster_center in cluster_centers:
        print(f"the distance between point {pt_name} {pt} and cluster {cluster_center} is:\n",
              f"    sqrt(({cluster_center[0]}-{pt[0]})**2 + ({cluster_center[1]}-{pt[1]})**2 =",
              f"{np.sqrt( (cluster_center[0]-pt[0])**2 + (cluster_center[1]-pt[1])**2 ):.5f}")
    print()
```
