```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.array([7.2, 1.4, 3.7, 4.8, 4.1, 6.1, 8.1, 5.6, 7.3, 8.2])
Y_train = np.array([0.8, 1.2, 0.6,-0.5,-0.1, 0.4, 1.6, 0.3, 1.9, 1.4])
X_test  = np.array([7.1, 1.2, 3.8, 4.6, 4.1, 6.5, 8.0, 5.4, 7.2, 8.1])
Y_test  = np.array([0.7, 1.2, 0.5,-0.5,-0.2, 0.4, 1.6, 0.4, 1.8, 1.2])
```

```python
"""(a)"""
'''cal theta'''
X_1 = np.concatenate((np.ones((10, 1)), X_train.reshape(10, 1)), axis=1)
# print(X_1)
Y = Y_train.reshape((10,1))
# print(Y)
Theta_head_1 = np.linalg.inv(X_1.T @ X_1) @ X_1.T @ Y
print('Theta_head_1:\n', Theta_head_1)

'''cal_predict'''
f_theta_x_1 = lambda x: Theta_head_1[0, 0] + Theta_head_1[1,0] * x
xs = np.linspace(0, 10, 100)
ys = f_theta_x_1(xs)

'''plot'''
plt.figure()
plt.scatter(X_train, Y_train, color='blue')
plt.plot(xs, ys, label='f1', color='red')
plt.title("1st order prediction")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
"""(b)"""
'''cal theta'''
X_2 = np.concatenate((np.ones((10, 1)), X_train.reshape(10, 1), (X_train**2).reshape(10, 1)), axis=1)
# print(X_2)
Y = Y_train.reshape((10,1))
# # print(Y)
Theta_head_2 = np.linalg.inv(X_2.T @ X_2) @ X_2.T @ Y
print('Theta_head_2:\n', Theta_head_2)

'''cal_predict'''
f_theta_x_2 = lambda x: Theta_head_2[0, 0] + Theta_head_2[1,0] * x + Theta_head_2[2, 0] * x**2
xs = np.linspace(0, 10, 100)
ys = f_theta_x_2(xs)

'''plot'''
plt.figure()
plt.scatter(X_train, Y_train, color='blue')
plt.plot(xs, ys, label='f1', color='red')
plt.title("1st order prediction")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
"""(c)"""
'''cal theta'''
X_4 = np.concatenate((np.ones((10, 1)), X_train.reshape(10, 1), (X_train**2).reshape(10, 1), (X_train**3).reshape(10, 1), (X_train**4).reshape(10, 1)), axis=1)
# print(X_4)
Y = Y_train.reshape((10,1))
# # print(Y)
Theta_head_4= np.linalg.inv(X_4.T @ X_4) @ X_4.T @ Y
print('Theta_head_3:\n', Theta_head_4)

'''cal_predict'''
f_theta_x_4 = lambda x: Theta_head_4[0, 0] + Theta_head_4[1,0] * x + Theta_head_4[2, 0] * x**2 + Theta_head_4[3, 0] * x**3 + Theta_head_4[4, 0] * x**4
xs = np.linspace(0, 10, 100)
ys = f_theta_x_4(xs)

'''plot'''
plt.figure()
plt.scatter(X_train, Y_train, color='blue')
plt.plot(xs, ys, label='f1', color='red')
plt.title("1st order prediction")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
