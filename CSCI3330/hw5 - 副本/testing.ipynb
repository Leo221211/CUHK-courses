```python
def renderLambertianImage(albedoIm, normalIm, lightVec):
    image = 0

    print(normalIm[0, 0])
    print(lightVec[0, 0])
    print(normalIm[0, 0] * lightVec[0, 0])

    ### YOUR CODE HERE
    image = np.zeros(albedoIm.shape)
    # image = normalIm @ lightVec * albedoIm
    for i in range(albedoIm.shape[0]):
        for j in range(albedoIm.shape[1]):
            image[i, j] = normalIm[i, j] * lightVec[0, 0]

    image = image * albedoIm
    ### END YOUR CODE

    return image

# surface normal
bunny_sn = np.load('data/bunny.npy')

# albedo
h, w, _ = bunny_sn.shape
bunny_albedo = np.ones((h, w))

# source 
light_sources = [np.array([0, 0, 1]).reshape(1,1,3), np.array([0, 1, 0]).reshape(1,1,3), np.array([1, 0, 0]).reshape(1,1,3)]


# Render the images under different light sources
rendered_images = []
for s in light_sources:
    rendered_image = renderLambertianImage(bunny_albedo, bunny_sn, s)
    rendered_images.append(rendered_image)  
```
