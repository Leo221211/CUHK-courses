### 3. Write math equations using *LaTex*

#### Basics

To embed LaTeX within text, simply encapsulate the LaTeX portions in dollar signs (`$`). MathJax takes care of the rest. As an example, consider the sentence below and the markdown/LaTeX code to render it.

Einstein told us that $E = mc^2$.

    Einstein told us that $E = mc^2$.

Notice how the equation is properly rendered, with mathematical variables in italics. Note also how `^2` was used to raise to a power. If the power has more than one character in it, it should be enclosed in braces (`{}`).  In fact, braces are used to generally group symbols in LaTeX.

Euler told us that $\mathrm{e}^{i \pi} - 1 = 0$.

    Euler told us that $\mathrm{e}^{i \pi} - 1 = 0$.

Aside from the grouping braces, there are several other syntactical items of note. First, notice that I made the special character $\pi$ with `\pi`. In general, a backward slash precedes special symbols or commands in LaTeX. If we want another Greek letter, like $\theta$, we use `\theta`. Now, also note that I used "`\mathrm{e}`" for the base of the natural logarithm.  I was signaling to LaTeX that I wanted the character written in Roman font, and not italics, so I used `\mathrm`. Anything in the braces following the function `\mathrm` is rendered in Roman font. Note the difference.

This is $e$. This is $\mathrm{e}$

    This is $e$. This is $\mathrm{e}$.

Now, back to grouping things in braces. We can do similar groupings using braces with with subscripts.

The dot product of two $n$-vectors is $\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i$.

    The dot product of two $n$-vectors is $\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i$.

Here, I have used `$\mathbf{a}$` to make the character `a` boldface, denoting a vector. Note that we denote subscripts with an underscore. Notice also that the bounds of the sum use the same underscore and caret notation as for subscripts and superscripts.

#### Command *align*: Displaying equations on separate lines

The bounds on the summation in the above example may look a little funny to you because they are not above and below the summation symbol.  This is because this particular equation is written inline.  If we had separated it from the text, it renders differently.

We can make an equation appear centered on a new line, like

\begin{align}
\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i.
\end{align}

The `align` environment in LaTeX specifies that you want centered equations, separated from the text. It is called `align` because it allows you to align the equations. You separate lines in the equations with a double backslash (`//`).  Insert an ampersand (`&`) in each line at the alignment point. All equations will be aligned at the location of the ampersand symbols (and, of course, the ampersands will not appear in the rendered equations).

For a three-vector consisting of $x$, $y$, and $z$ components,

\begin{align}
\mathbf{a} \cdot \mathbf{b} &= \sum_{i=1}^n a_i b_i \\
&= a_x b_x + a_y b_y + a_z b_z.
\end{align}

## II. Introdution to Numpy

### Basics of Array
Similar to `list` of python (supports index, slice, etc.).

`````python
import numpy as np

a = np.array([1, 2, 3])   # Create a 1-dimensional array
print(type(a))            # Prints "<class 'numpy.ndarray'>"
print(a.shape)            # Prints "(3,)"
print(a[0], a[1], a[2])   # Prints "1 2 3"
a[0] = 5                  # Modify an element of the array
print(a)                  # Prints "[5, 2, 3]"

b = np.array([[1,2,3],[4,5,6]])    # Create a 2-dimensional array
print(b.shape)                     # Prints "(2, 3)"
print(b[0, 0], b[0, 1], b[1, 0])   # Prints "1 2 4"
`````

Numpy also provides some useful functions to quickly build all-zero, all-one, all-constant, identity matrix, and random array.

`````python
arr_all_zero = np.zeros((2,2))   # Create an array of all zeros
print(f"arr_all_zero=\n{arr_all_zero}\n")

arr_all_one = np.ones((1,2))    # Create an array of all ones
print(f"arr_all_one=\n{arr_all_one}\n")

arr_all_constant = np.full((2,2), 7)  # Create a constant array
print(f"arr_all_constant=\n{arr_all_constant}\n")

arr_identity = np.eye(2)         # Create a 2x2 identity matrix
print(f"arr_identity=\n{arr_identity}\n")

arr_random = np.random.random((2,2))  # Create an array filled with random values
print(f"arr_random=\n{arr_random}\n")
`````

### Calculation with Arrays
These array can be involved in all basic python arithmetic operators (`+, -, * , /`, etc.) and some calculation in linear algera.

`````python
x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)

# Elementwise sum; both produce the array
# [[ 6.0  8.0]
#  [10.0 12.0]]
print(x + y)
print(np.add(x, y))

# Elementwise difference; both produce the array
# [[-4.0 -4.0]
#  [-4.0 -4.0]]
print(x - y)
print(np.subtract(x, y))

# Elementwise product; both produce the array
# [[ 5.0 12.0]
#  [21.0 32.0]]
print(x * y)
print(np.multiply(x, y))

# Elementwise division; both produce the array
# [[ 0.2         0.33333333]
#  [ 0.42857143  0.5       ]]
print(x / y)
print(np.divide(x, y))

# Elementwise square root; produces the array
# [[ 1.          1.41421356]
#  [ 1.73205081  2.        ]]
print(np.sqrt(x))
`````

Numpy also support linear algebra operations on arrays (matrixs).
⚠️ Note: Operator `*` for two 2-d arrays (matrixs) are element-wise multiplication, instead of matrix multiplication.

Use `np.dot()` for matrix multiplication.


`````python
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])

v = np.array([9,10])
w = np.array([11, 12])

print(x,y,v,w)

# Inner product of vectors; both produce 219
print(v.dot(w))
print(np.dot(v, w))

# Matrix / vector product; both produce the rank 1 array [29 67]
print(x.dot(v))
print(np.dot(x, v))

# Matrix / matrix product; both produce the rank 2 array
# [[19 22]
#  [43 50]]
print(x.dot(y))
print(np.dot(x, y))
`````

Numpy also provides some reducing operations like sum, mean, min, and max.

Remember to give the dimension/axis (starting from 0) that you want to reduce.

If no axis is given, the array will be reduced to a scalar.

`````python
x = np.array([[1,2],[3,4]])
print('np.sum():')
print(np.sum(x))  # Compute sum of all elements
print(np.sum(x, axis=0))  # Compute sum of each column
print(np.sum(x, axis=1))  # Compute sum of each row

print('\n np.mean():')
print(np.mean(x))  # Compute mean of all elements
print(np.mean(x, axis=0))  # Compute mean of each column
print(np.mean(x, axis=1))  # Compute mean of each row

print('\n np.min():')
print(np.min(x))  # Compute minimum of all elements
print(np.min(x, axis=0))  # Compute minimum of each column
print(np.min(x, axis=1))  # Compute minimum of each row

print('\n np.max():')
print(np.max(x))  # Compute maximum of all elements
print(np.max(x, axis=0))  # Compute maximum of each column
print(np.max(x, axis=1))  # Compute maximum of each row
`````

Arrays can be reshaped to any shape you want, as the new shape can hold that same number of elements in current array.

Create an array $w$ with shape (1, 6):

`````python
w = np.array([[4,5,6,7,8,9]])    # w has shape (1, 6)
print(f'w=\n{w}')
print(f"w.shape={w.shape}")
`````

Let reshape $w$ to (2, 3).

`````python
w = np.reshape(w, (2, 3))
print(f'w=\n{w}')
print(f"w.shape={w.shape}")
`````

Transpose the $w$ and calculate the gram matrix of $w$, i.e., $w^T w$

`````python
print(f'w=\n{w}')
wT = w.T
print(f'wT=\n{wT}')
wTw = np.dot(wT, w)
print(f'wTw=\n{wTw}')
print(f'wTw.shape={wTw.shape}')
`````

## III. Introdution to Matplotlib

### Basics

We first import the `matplotlib`'s subpackage `pyplot` as `plt`.

`````python
import matplotlib.pyplot as plt
`````

Then, we generate a seris of discrect point on the x-axis using `np.linspace` as the ticks of our following figures.

And plot sine and cosine function with the x-range of [0, 10], containing 100 sampling points for each line.

`````python
import numpy as np
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
`````

We can customize the style of the line in the plot by adding extra arguments.

For example, `'--'` stands for dashed line.

`````python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
`````

### Sub-Plot

The `plt` object maintain a "hidden figure" after you call `plt.figure()`.

All operations afterwards will be applied on this figure.

Create a subplot, a parition of the original `figure`, using `plt.subplot(rows, columns, current_panel_number)`.

Then we show an example of plotting the sine and cosine function seperately in two sub-plots.

`````python
plt.figure()  # create a plot figure

# # create the first of two panels and set current axis
# plt.subplot(2, 1, 1) # (rows, columns, panel number)
# plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
`````

You can edit each sub-plot's style, axis limit, ticks, etc. as you want.

Each sub-plot's settings are independent from others' settings.

Here are an example

`````python
plt.figure()  # create a plot figure

plt.subplot(2, 1, 1) # (rows, columns, panel number), create the first of two panels and set current axis
# [!] context of sub-plot 1 starts
plt.plot(x, np.sin(x), color='#FF0000')
plt.xlim(0, 15)
plt.ylim(-1.0, 0.5)

plt.subplot(2, 1, 2) # create the second panel and set current axis
# [!] context of sub-plot 1 ends, and context of sub-plot 1 starts
plt.xlim(0, 5)
plt.ylim(-0.5, 1.0)

plt.plot(x, np.cos(x), '--', color='#333333');
`````

We can even draw animation with matplotlib!

The keyframe function `animate(t)` draw each frame of our animation according to the given timestep `t`.

`````python
import matplotlib.animation
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['animation.html'] = "jshtml"
plt.rcParams['figure.dpi'] = 100
plt.ioff()
fig, ax = plt.subplots()

def animate(t):
    plt.cla()
    plt.plot(x, np.sin(0.05*t*x))
    plt.xlim(0, 10)
    plt.ylim(-1.0, 1.0)
    plt.tight_layout() # Make the padding of the figue smaller

matplotlib.animation.FuncAnimation(fig, animate, frames=20)
`````
