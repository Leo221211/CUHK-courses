```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
# full version
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

x = np.linspace(0, 10, 100)

ax.plot(x, np.cos(x))

fig.show()
```

```python
# simplified
plt.figure()

plt.title("new single axis")

plt.plot(x, np.sin(x))
```

```python
# simplified multiple axes
plt.figure()

plt.subplot(3, 2, 1)

plt.title("ax 1")

plt.plot(x, np.sin(x))

plt.subplot(3, 2, 2)

plt.title("ax 2")

plt.plot(x, np.cos(x))

plt.subplot(3, 2, 3)

plt.title("ax 3")

plt.plot(x, np.sqrt(x))
```
